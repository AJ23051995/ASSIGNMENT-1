Question 2-In the above question, now scrape the following details of each product listed in first 3 pages of your 
search results and save it in a data frame and csv. In case if any product has less than 3 pages in search 
results then scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“

```python
# Defining empty arrays for the data to be scraped
brand = []
name = []
prc = []
ret = []
deli = []
avail = []
purl = []
```

```python
# Forming a loop to scrape all the data required
for url in link:
    driver.get(url) #Calling each url one-by-one
    
    #Scraping brand name
    try:
        tag1 = driver.find_element(By.ID,'productTitle')
        brand.append(tag1.text.split()[0])
    except:
        brand.append("-")
    
    #Scraping Product Name
    try:
        tag2 = driver.find_element(By.ID,'productTitle')
        name.append(tag2.text)
    except:
        name.append("-")
        
    #Scraping Price
    try:
        tag3 = driver.find_element(By.XPATH,'//span[@class="a-price a-text-price a-size-medium apexPriceToPay"]/span[2]')
        prc.append(tag3.text)
    except:
        prc.append("-")
        
    #Scraping Return/Exchange
    try:
        tag4 = driver.find_element(By.ID,'RETURNS_POLICY')
        ret.append(tag4.text)
    except:
        ret.append("-")
        
    #Scraping Delivery 
    try:
        tag5 = driver.find_element(By.XPATH,'//div[@id="ddmDeliveryMessage"]/b[1]')
        deli.append(tag5.text)
    except:
        deli.append("-")
        
    #Scraping Availability
    try:
        tag6 = driver.find_element(By.XPATH,'//div[@id="availability"]/span[1]')
        avail.append(tag6.text)
    except:
        avail.append("-")
    
    #scrapping product url
    purl.append(url)
```

```python
#Creating a DataFrame of the scraped data
df = pd.DataFrame({"Brand Name":brand, 
                   "Name of the Product":name, 
                   "Price":prc,
                   "Return/Exchange":ret,
                   "Expected Delivery":deli,
                   "Availability":avail,
                   "Product URL":purl})
```

scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”,

```python
#Creating an empty array to store the link for the products
link = []
```

```python
#Scraping the link to all the products on page 1
tag = driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
for i in tag:
    link.append(i.get_attribute('href'))
```

```python
len(link)
```

```python
# Defining empty arrays for the data to be scraped
brand = []
name = []
clr = []
ram = []
rom = []
pcm = []
scm = []
dis = []
btt = []
prc = []
purl = []
```

```python
# Forming a loop to scrape all the data required
for url in link:
    driver.get(url) #Calling each url one-by-one
    
    #Scraping brand name
    try:
        tag1 = driver.find_element(By.XPATH,'//span[@class="B_NuCI"]')
        brand.append(tag1.text.split()[0])
    except:
        brand.append("-")
    
    #Scraping Product Name
    try:
        tag2 = driver.find_element(By.XPATH,'//span[@class="B_NuCI"]')
        name.append(tag2.text.split("(")[0])
    except:
        name.append("-")
        
    #Scraping Colour
    try:                                     
        tag3 = driver.find_element(By.XPATH,'//span[@class="B_NuCI"]')
        clr.append(tag3.text.split("(")[1].split(",")[0])
    except:
        clr.append("-")
        
    #Scraping RAM
    try:
        tag4 = driver.find_element(By.XPATH,'//span[@class="B_NuCI"]')
        ram.append(tag4.text.split("(")[-1].split(")")[0])
    except:
        ram.append("-")
    
    #Scraping ROM
    try:
        tag5 = driver.find_element(By.XPATH,'//span[@class="B_NuCI"]')
        rom.append(tag5.text.split("(")[1].split(",")[1].split(")")[0])
    except:
        rom.append("-")
        
    #Scraping Primary Camera
    try:
        tag6 = driver.find_element(By.XPATH,'//div[@class="_2418kt"]/ul/li[3]')
        pcm.append(tag6.text.split("|")[0])
    except:
        pcm.append("-")
        
    #Scraping Secondary Camera
    try:
        tag7 = driver.find_element_by(By.XPATH,'//div[@class="_2418kt"]/ul/li[3]')
        scm.append(tag7.text.split("|")[1])
    except:
        scm.append("-")
        
    #Scraping Display
    try:
        tag8 = driver.find_element(By.XPATH,'//div[@class="_2418kt"]/ul/li[2]')
        dis.append(tag8.text)
    except:
        dis.append("-")
        
    #Scraping Battery
    try:
        tag9 = driver.find_element(By.XPATH,'//div[@class="_2418kt"]/ul/li[4]')
        btt.append(tag9.text)
    except:
        btt.append("-")
        
    #Scraping Price
    try:
        tag10 = driver.find_element(By.XPATH,'//div[@class="_30jeq3 _16Jk6d"]')
        prc.append(tag10.text)
    except:
        prc.append("-")
        
    #Scraping Product URL
    purl.append(url)
```

```python
#Creating a DataFrame of the scraped data
df = pd.DataFrame({"Brand Name":brand, 
                   "Smartphone Name":name,
                   "Colour":clr,
                   "RAM":ram,
                   "Storage(ROM)":rom,
                   "Primary Camera":pcm,
                   "Secondary Camera":scm,
                   "Display Size":dis,
                   "Battery Capacity":btt,
                   "Price":prc,
                   "Product URL":purl})
```
